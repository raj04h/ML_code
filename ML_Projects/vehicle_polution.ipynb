```python
from sklearn.tree import DecisionTreeClassifier
```

```python
x=fuel_encoded.drop("FUELTYPE",axis=1)
y=fuel_encoded["FUELTYPE"]
```

```python
model=DecisionTreeClassifier()
model.fit(x,y)

feature_import=pd.DataFrame({"Feature": x.columns,"Importance":model.feature_importances_})
feature_import=feature_import.sort_values(by="Importance",ascending=False)
print(feature_import)
```

```python
pyplt.figure(figsize=(10,6))
sb.barplot(x="Importance",y="Feature",data=feature_import)
pyplt.title("Feature Importance")
pyplt.show()
```

```python
top_feature=feature_import["Feature"].head(5)
x_selected=x[top_feature]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, mean_squared_error,precision_score,recall_score,f1_score,roc_auc_score,roc_curve,auc

from sklearn.preprocessing import label_binarize
```

```python
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
y_pred=clf.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
precission=precision_score(y_test,y_pred,average="weighted")
recall=recall_score(y_test,y_pred,average="weighted")
f1=f1_score(y_test,y_pred,average="weighted")
```

```python
y_test_bin = label_binarize(y_test, classes=[0, 1, 2, 3, 4])  
y_prob = clf.predict_proba(x_test)
roc_auc = roc_auc_score(y_test_bin, y_prob, multi_class='ovr')
```

```python
print(f"Accuracy: {accuracy}")
print(f"Precision: {precission}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")
print(f"ROC-AUC Score: {roc_auc}")
```

```python
print("y_test_bin shape:", y_test_bin.shape)
print("y_prob shape:", y_prob.shape)
```

```python
num_classes = min(y_test_bin.shape[1], y_prob.shape[1])
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc_value[i] = auc(fpr[i], tpr[i])
```

```python
# Assuming y_test_bin and y_prob have been defined and contain the correct shapes.
fpr = {}
tpr = {}
roc_auc_value = {}
```

```python
# Calculate ROC curve and AUC for each class
num_classes = min(y_test_bin.shape[1], y_prob.shape[1])
for i in range(num_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc_value[i] = auc(fpr[i], tpr[i])
```

```python
# Plot ROC curves
pyplt.figure(figsize=(10, 8))
for i in range(num_classes):
    pyplt.plot(fpr[i], tpr[i], label=f'Class {i} (AUC = {roc_auc_value[i]:.2f})')

# Plotting the diagonal line
pyplt.plot([0, 1], [0, 1], 'k--', label='Random Guessing')

# Adding labels and legend
pyplt.xlabel('False Positive Rate')
pyplt.ylabel('True Positive Rate')
pyplt.title('Receiver Operating Characteristic (ROC) Curve')
pyplt.legend(loc='lower right')
pyplt.grid()
pyplt.show()
```
